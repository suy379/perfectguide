```python
ti_df = pd.read_csv('01/titanic_train.csv')
ti_df.head()
```

```python
#df.apply(lambda)

#한번 name 컬럼의 글자수 세는 새로운 컬럼을 만들어보자.
ti_df['Name_len'] = ti_df['Name'].apply(lambda x : len(x))
ti_df.head()
```

```python
#lambda 식에 if else 절 사용해보기->항상 출력문을 앞에쓰고 if 조건문, else .. 

#나이가 15세 미만이면 'Child', 그렇지 않으면 'Adult'로 구분하는 새로운 컬럼을 만들어보자.
ti_df['Child_Adult'] = ti_df['Age'].apply(lambda x : 'Child' if x<15 else 'Adult')
ti_df.head(10)
```

```python
#심화- else if를 쓰고싶다면 else절을 ( )로 묶고 그 안에 다시 if else를 적용
#나이가 15세 이하면 Child, 15-60 사이는 Adult, 61세 이상은 Elderly로 분류
ti_df['Age_Cat'] = ti_df['Age'].apply(lambda x : 'Child' if x<=15 else ('Adult' if x<61 else 'Elderly'))
ti_df.tail()
```

```python
#복잡하다면 def 함수 생성하는게 더 좋을수도 있다.

def get_category(age):
    cat= ''
    if age <=5 : cat='Baby'
    elif age <=12 : cat = 'Child'
    elif age <=18 : cat = 'Teenager'
    elif age <=25: cat = 'Student'
    elif age <=35: cat = 'Yonung Adult'
    elif age <=60: cat = 'Adult'
    else : cat='Elderly'
        
    return cat
```

```python
ti_df['Age_Cat2']= ti_df['Age'].apply(get_category)
ti_df.head()
```
